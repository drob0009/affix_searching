```python
from ExileAPI import ExileAPI
import logging
import pandas as pd

pd.set_option('display.max_colwidth', None)
api = ExileAPI(log_level=logging.ERROR)
```

```python
tags = set()
td = dict()
for affix in api.affixes:
    if affix.craftable:
        tags |= affix.as_dict()["tags"]
df = pd.DataFrame([{"tag": tag} for tag in tags])
df.sort_values(by=["tag"])
```

```python
find_by_tag = set([
    affix
    for affix in api.find_affix_by_tag(["chaos"])
    if affix.craftable and "minion" not in affix.tags and "caster" not in affix.tags
])

find_by_slot = set([
    affix
    for affix in api.find_affix_by_type(["jewel"], True)
    if affix.craftable
])

results = find_by_tag.intersection(find_by_slot)

results = ExileAPI.as_df(results)
results[["desc", "type", "tags", "domain", "base_types", "essence", "influence"]].sort_values(by=["type", "domain"])
```
