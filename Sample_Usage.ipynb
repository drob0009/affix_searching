```python
from ExileAPI import ExileAPI
import logging
import pandas as pd

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
api = ExileAPI(log_level=logging.DEBUG)
```

# Affix by class

These are the various item tags we can search for affixes by.
For example 'dex_armour' would show us affixes specific to dex armor pieces.

as well as the more general classes (e.g. Body Armor)

```python
for k, v in api.affixes_by_class.items():
    print(f"{k}:\n\t{v}")
```

```python
amulet_affixes = set(api.find_affix_by_class(["Amulet"]))
chaos_affixes = set(api.find_affix_by_tag(["chaos"]))
ExileAPI.as_df(list(amulet_affixes & chaos_affixes))[["desc", "name", "type", "tags", "domain", "item_affix_tags", "essence", "influence"]].sort_values(by=["type", "domain", "influence"])
```
